# Experimenting with ResNet50

```python
# Imports modules, dataset and results paths
from modules.paths import (
    TESTING_DATASET,
    TRAINING_DATASET,
)
from models.trainer import (
    build_model,
    run_experiment,
)
from models.metrics import display_results, load_results
from models.plots import plot_history
```

```python
# Running ResNet50 Experiment with Data Augmentation and Normalization & Batch 32
resnet_results = run_experiment(
    name="resnet50",
    model_builder=lambda: build_model("resnet50"),
    train_dir=TRAINING_DATASET,
    val_dir=TESTING_DATASET,
    test_dir=TESTING_DATASET,
    image_size=224,
    batch_size=32,
    epochs=50,
    lr=5e-4,
    augment=True,
    normalize=True,
)
```

```python
# Displaying results if available in notebook variables and outputs
display_results("ResNet50", resnet_results)
```

```python
# Displaying learning and validation curves if available in notebook variables and outputs
plot_history(
    resnet_results["history"],
    "ResNet50 (batch 32) Training and Validation Accuracy",
)
```
