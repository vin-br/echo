# Experimenting with ConvNeXt Tiny

```python
# Imports modules, dataset and results paths
from modules.paths import (
    TESTING_DATASET,
    TRAINING_DATASET,
)
from models.trainer import (
    build_model,
    run_experiment,
)
from models.metrics import display_results, load_results
from models.plots import plot_history
```

## ConvNeXt Tiny Batch 16

```python
# Running ConvNeXt Tiny Experiment with Data Augmentation and Normalization & Batch 16
convnext_results_b16 = run_experiment(
    name="convnext-tiny",
    model_builder=lambda: build_model("convnext-tiny"),
    train_dir=TRAINING_DATASET,
    val_dir=TESTING_DATASET,
    test_dir=TESTING_DATASET,
    image_size=224,
    batch_size=16,
    epochs=60,
    lr=5e-4,
    augment=True,
    normalize=True,
)
```

```python
# Displaying results if available in notebook variables and outputs
display_results("ConvNeXt-Tiny", convnext_results_b16)
```

```python
# Displaying learning and validation curves if available in notebook variables and outputs
plot_history(
    convnext_results_b16["history"],
    "ConvNeXt-Tiny (batch 16) Training and Validation Accuracy",
)
```

```python
# Loading results from JSON file
convnext_b16_json_results = load_results("convnext-tiny-b16-img224-e60-lr-5e-4.json")

# Displaying results if available in results as JSON file
display_results("ConvNeXt-Tiny (saved)", convnext_b16_json_results)
```

```python
# Displaying learning and validation curves if available in notebook variables and outputs
plot_history(
    convnext_b16_json_results["history"],
    "ConvNeXt-Tiny Training and Validation Accuracy from JSON Results",
)
```
