```python
# Imports modules, dataset and results paths
from modules.paths import (
    TESTING_DATASET,
    TRAINING_DATASET,
)
from models.trainer import (
    build_model,
    run_experiment,
)
from models.metrics import display_results, load_results
from models.plots import plot_history
```

### MobileNetV3-Large (batch 32)
Smaller backbone better suited for CPU-only fine-tuning.

```python
# Running MobilenetV3 Large Experiment with Data Augmentation and Normalization & Batch 32
mobilenet_results = run_experiment(
    name="mobilenet-v3-large",
    model_builder=lambda: build_model("mobilenet-v3-large"),
    train_dir=TRAINING_DATASET,
    val_dir=TESTING_DATASET,
    test_dir=TESTING_DATASET,
    image_size=224,
    batch_size=32,
    epochs=60,
    lr=5e-4,
    augment=True,
    normalize=True,
)
```

```python
# Displaying results if available in notebook variables and outputs
display_results("MobileNetV3-Large", mobilenet_results)
```

```python
# Displaying learning and validation curves if available in notebook variables and outputs
plot_history(
    mobilenet_results["history"],
    "MobileNetV3-Large Training and Validation Accuracy",
)
```
